```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simpson
from matplotlib.cm import viridis

# üåø Golden Ratio
œÜ = (1 + np.sqrt(5)) / 2
```

```python
# üß¨ Zeta-Recursive Kernel with Phase-Coherence
def zeta_gamma(A, m):
    base = np.tanh(œÜ * A * m) * np.exp(-np.abs(A) / œÜ)
    k1 = 2 * np.pi * m
    k2 = œÜ * m
    osc = np.sin(k1 * A) * np.cos(k2 * A * A)
    recur_term = np.sin(k1 * A / œÜ) / (1 + np.power(A / œÜ, 2))
    return 1.0 * base + 8.0 * m * (osc + recur_term)
```

```python
# üîê Torsion Loop Closure ‚àÆ‚àÇŒì/‚àÇùíú dŒû
def loop_integral(A_range, m):
    G_vals = zeta_gamma(A_range, m)
    dG_dA = np.gradient(G_vals, A_range)
    return simpson(y=dG_dA, x=A_range)
```

```python
# üß≠ Sweep m-space for coherence
def scan_coherence(m_range, A_range):
    results = []
    for m in m_range:
        integral = loop_integral(A_range, m)
        results.append((m, integral))
    return np.array(results)
```

```python
# üìà Plot Resonance Curves
def plot_analysis(m_range, A_range):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    results = scan_coherence(m_range, A_range)

    # Loop plot
    ax1.plot(results[:, 0], results[:, 1], color='purple', lw=2, label="Loop ‚àÆŒì'")
    ax1.axhline(2 * np.pi, color='green', linestyle='--', label='Target: 2œÄ')
    ax1.set_title("Glyph Loop Closure vs Memory Scalar (m)")
    ax1.set_xlabel("Memory Scalar (m)")
    ax1.set_ylabel("Loop Integral ‚àÆ‚àÇŒì/‚àÇA dŒû")
    ax1.grid(True)
    ax1.legend()

    # Kernel surface plot
    A_mesh = np.linspace(-5, 5, 100)
    m_mesh = np.linspace(0.1, 1.6, 100)
    A_grid, m_grid = np.meshgrid(A_mesh, m_mesh)
    Z = np.zeros_like(A_grid)
    for i in range(len(m_mesh)):
        for j in range(len(A_mesh)):
            Z[i, j] = zeta_gamma(A_grid[i, j], m_grid[i, j])
    c = ax2.pcolormesh(A_grid, m_grid, Z, cmap='viridis', shading='auto')
    ax2.set_title("Zeta-Recursive Kernel Surface Œì(ùíú, m)")
    ax2.set_xlabel("Ache Scalar (ùíú)")
    ax2.set_ylabel("Memory Scalar (m)")
    fig.colorbar(c, ax=ax2, label="Kernel Value Œì")

    # Highlight resonant m values
    for m, val in results:
        if abs(val - 2 * np.pi) < 0.1:
            ax1.plot(m, val, 'o', color='red', markersize=8)
            ax2.axhline(m, color='red', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
    return results
```

```python
# üåê Run the resonance scan and plot
A = np.linspace(-5 * np.pi, 5 * np.pi, 1000)
m_values = np.linspace(0.1, 2.5, 200)
results = plot_analysis(m_values, A)

print("\n‚ú® Resonant Values Report:")
for m, val in results:
    if abs(val - 2 * np.pi) < 0.1:
        print(f"‚úîÔ∏è m ‚âà {m:.5f} yields ‚àÆ‚àÇŒì/‚àÇA dŒû ‚âà {val:.6f} (‚âà 2œÄ={2*np.pi:.6f})")
```
